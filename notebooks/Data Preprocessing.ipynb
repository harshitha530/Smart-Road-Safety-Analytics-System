```python
df.isnull().sum()
```

```python
df['festival'] = df['festival'].fillna('No Festival')
```

```python
df['date'] = pd.to_datetime(df['date'])

df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month
df['day'] = df['date'].dt.day

df.drop('date', axis=1, inplace=True)
```

```python
df.head()
```

```python
df.drop('time', axis=1, inplace=True)
df.drop('accident_id', axis=1, inplace=True)
df.drop('is_weekend',axis=1,inplace=True)
```

```python
df.to_csv("../data/road_safety_visualization.csv", index=False)
```

```python
print(df.columns.tolist())
```

```python
df_ml = df.copy()
```

```python
from sklearn.preprocessing import LabelEncoder

categorical_cols = [
    'city','state','road_type',
    'traffic_signal','weather','visibility',
    'traffic_density','cause',
    'accident_severity','festival','day_of_week'
]

for col in categorical_cols:
    le = LabelEncoder()
    le.fit(df_ml[col])
    
    print(f"\n{col}")
    for i, value in enumerate(le.classes_):
        print(f"{value} -> {i}")
```

```python
for col in categorical_cols:
    df_ml[col] = le.fit_transform(df_ml[col])
```
